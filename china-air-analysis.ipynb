```r
a <- "../input/pm25-data-for-five-chinese-cities/BeijingPM20100101_20151231.csv"
b <- "../input/pm25-data-for-five-chinese-cities/GuangzhouPM20100101_20151231.csv"
c <- "../input/pm25-data-for-five-chinese-cities/ChengduPM20100101_20151231.csv"
d <- "../input/pm25-data-for-five-chinese-cities/ShanghaiPM20100101_20151231.csv"
e <- "../input/pm25-data-for-five-chinese-cities/ShenyangPM20100101_20151231.csv"
```

```r
aa <- read_csv(a)
bb <- read_csv(b)
cc <- read_csv(c)
dd <- read_csv(d)
ee <- read_csv(e)
aa$name <- "Bejing"
aa
```

```r
aa$name <- "Bejing"
bb$name <- "Chengdu"
cc$name <- "Guangzhou"
dd$name <- "Shanghai"
ee$name <- "Shenyang"
```

```r
commcol <- intersect(colnames(aa), colnames(bb))
pr <- rbind(
  aa[commcol], bb[commcol], cc[commcol], dd[commcol], ee[commcol]    
)
head(pr,  row = 20)
```

```r
#Convert to data frame 
pr <- as.data.frame(pr)
pr1 <- pr %>% select_if(is.numeric)

p1 <- principal(pr1, nfactors = 4, covar = TRUE)
print(p1)

pc <- cor(pr1, use = "complete.obs")
pc <- as.data.frame(pc)
pc
```

```r
#Regression of non numeric Date to predict pm2.5
m3 <- lm(pr$`PM_US Post` ~ pr$year + pr$Iws + pr$TEMP + pr$cbwd)
summary(m3)

m4 <- lm(pr$`PM_US Post` ~ pr$season + pr$year)
summary(m4)
```
